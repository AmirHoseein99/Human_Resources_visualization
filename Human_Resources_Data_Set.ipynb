```python
plt.figure(figsize=(12, 5))
sns.histplot(dataset.DeptID.values)
```

```python
plt.figure(figsize=(8, 8))
dataset['Sex'].value_counts().plot(kind='pie',autopct='%1.1f%%');
plt.title('Employees gender');
```

```python
plt.figure(figsize = [20, 5])
ordered = dataset['TermReason'].value_counts().index[1:]
sns.countplot(data=dataset, x= 'TermReason', order = ordered)
plt.title('Employment Status')
plt.xticks(rotation=45)
```


Is there any relationship between who a person works for and their performance score?

```python
plt.figure(figsize = [15, 5])
Ord = dataset['ManagerName'].value_counts().index
sns.countplot(data= dataset, x='ManagerName', hue ='PerformanceScore', order=Ord)
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.title('The relationship between ManagerName & Performance Score')
```

```python
plt.figure(figsize = [15, 5])
Ord = dataset['RecruitmentSource'].value_counts().index
sns.countplot(data= dataset, x='RecruitmentSource', hue ='PerformanceScore', order=Ord)
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.title('The relationship between Recruitment Source & Performance Score')
```

The relationship between Department & Performance Score & Employees Satisfaction.

```python
means = dataset.groupby(['Department', 'PerformanceScore']).mean()['EmpSatisfaction']
means = means.reset_index(name = 'EmpSatisfaction_avg')
means = means.pivot(index = 'PerformanceScore', columns = 'Department',
                            values = 'EmpSatisfaction_avg')
plt.figure(figsize = [15, 5])    
sns.heatmap(means, annot = True, fmt = '.3f',
           cbar_kws = {'label' : 'mean(EmpSatisfaction)'});
plt.title('The relationship between Department & Performance Score & Employees Satisfaction.');
```
